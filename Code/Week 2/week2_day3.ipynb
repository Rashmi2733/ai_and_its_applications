### Exercise 6: Dependent Events

```python
#Lets calculate the probability of drawing two aces without replacement from a deck of 52 cards

total_cards = 52
ace = 4

prob_first_ace = ace / total_cards
prob_second_ace = (ace - 1) / (total_cards - 1)

prob = prob_first_ace * prob_second_ace

prob_with_replacement = prob_first_ace * prob_first_ace

print("Probability of drawing two aces without replacement is", prob)
print("Probability of drawing two aces with replacement is", prob_with_replacement)
```

```python
# Task 6: Calculate the probability of drawing two red cards without replace-
# ment from a deck (26 red cards, 52 total). Print the result.

total_cards = 52
red = 26

prob_first_red = red / total_cards
prob_second_red = (red - 1) / (total_cards - 1)

prob = prob_first_red * prob_second_red

prob_with_replacement = prob_first_red * prob_first_red

print("Probability of drawing two reds without replacement is", prob)
print("Probability of drawing two reds with replacement is", prob_with_replacement)
```
