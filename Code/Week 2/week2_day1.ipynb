# Measure of central tendency

## Exercise 1: Calculating the mean

```python
import pandas as pd
##Creating a panda series

data = pd.Series([2, 4, 6, 8, 10])
print("Original data:\n", data)

mean = data.mean()
print("Mean of the data:", int(mean))
```

### Exercise 2: Calculating the median

```python
median = data.median()
print("Median of the data:", median)
```

## Exercise 3: Calculating the mode

```python
data = pd.Series([2,5,8,8,8,4,5,8,10])
mode = data.mode()[0]
print("Mode of the data:", mode)
```

# Measure of variability

## Exercise 4: Calculating range

```python
data = pd.Series([2,4,5,8,10])

range = max(data) - min(data)
print("Range of the data is:", range)
```

## Exercise 5: Calculating variance and standard deviation

```python
data = pd.Series([2,4,5,8,10])

variance = data.var()
standard_deviation = data.std()

print("Variance of the data is:", variance)
print("Standard Deviation of the sata is:", standard_deviation)
```

```python
##Pie chart

import matplotlib.pyplot as plt

Names = ['Alice', 'Bob', 'Cathy',
            'Dan', 'Erika']

Scores = [34, 67, 22, 93, 87]

plt.pie(Scores,  autopct = '%1.1f%%')
plt.legend(labels = Names, loc = 'upper right')
plt.title("Pie Chart")

plt.show()
```

```python
Names = ['Alice', 'Bob', 'Cathy', 'Dan', 'Erika']

Scores = [34, 67, 22, 93, 87]

plt.pie(Scores,  autopct = '%1.1f%%', pctdistance=0.75)

# draw a white circle in the center of the pie chart which then creates a donut chart
centre_circle = plt.Circle((0, 0), radius=0.45, fc='white')

##Adding the circle we created above to our existing pie chart
fig = plt.gca().add_artist(centre_circle)

plt.legend(labels = Names, loc = 'upper right')

plt.title('Donut Chart')

plt.show()
```
