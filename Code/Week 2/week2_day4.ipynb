### Data Profiling and quality check

```python
##Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = sns.load_dataset('iris')
print(df.head(5))
print(df.tail())
```

```python
#Display basic information
print("Basic information:")
print(df.info())
```

```python
#Summary stats
print("Summary Statistics:")
print(df.describe())
```

```python
# Checking null values
print("Checking null values:")
print(df.isnull().sum())
```

```python
# Duplicated rows
print("Duplicated rows:")
print(df.duplicated().sum())
```

```python
#Value count a categorical or numerical column
print("Count of each species:")
print(df['species'].value_counts())
```

```python
print("Count of each petal length:")
print(df['petal_length'].value_counts())
```

```python
## Visualizing categorical variable

plt.figure(figsize= (6,8))
sns.countplot(x = 'species', data = df)
plt.title("Visualizing using seaborn")
plt.xlabel("Species")
plt.ylabel("Count of species")
```

```python
#Basic stats with functions

iris_numeric = df[['sepal_length', 'sepal_width', 'petal_width', 'petal_length']]
print("Mean of each column:")
print(iris_numeric.mean())

print("Median of each column:")
print(iris_numeric.median())

print("Mode of each column:")
print(iris_numeric.mode())
```

```python
## Pair plot
sns.pairplot(df, hue = 'species')
plt.show()
```

```python
##correlation matrix (numeric only)
corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title("Correlation Matrix")
plt.show()

##The highest correlation can be seen between petal length and petal width --- increasing petal length will increase petal width as well
##The lowest corrrelation can be seen between sepal width and sepal length --- increasing sepal length will decrease sepal width as well
```

```python
#Creating a boxplot
sns.boxplot(df)
plt.title("Boxplot for iris data")
plt.xlabel("Species")
plt.ylabel("Values (cm)")
plt.show()

##Sepal length column has the highest value (about 8 cm)
##Petal width column has the lowest value (about 0.1 cm)
##We see outliers in case of sepal_width
##The most variation is seen in petal length
##The least variance is seen in sepal width
```

```python
#Creating a histogram
sns.histplot(df['petal_length'], bins = 12)
```

```python
## The count of petals with length between 1.5 and 2 cm is the highest (about 26)
##there are no petal length values between 2 and 3 cm
#The lowest number of petal length values lie between 3 and 3.5 cm
```
