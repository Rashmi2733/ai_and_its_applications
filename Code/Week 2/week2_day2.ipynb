### Exercise 1: Working with scalars and vectors

```python
import numpy as np
```

```python
#Creating a scalar
scaler  = 7

#Creating a vector
vector = np.array([3,5,7])
print("Vector:", vector)

#Scaling vector through a scalar aka scaled vector
scaled_vector = scaler * vector
print("Scaled vector:", scaled_vector)
```

## Exercise 2: Addition and subtraction

```python
vector1 = np.array([3, 6, 12])
vector2 = np.array([5, 2, 15])

add_vector = vector1 + vector2
subtract_vector = vector1 - vector2

print("Added vector:", add_vector)
print("Subtracted vector:", subtract_vector)
```

```python
A = np.array([[1,3], [9,3]])
squared = np.dot(A, A)
print(squared)
```

```python
#Implementing matrix multiplication from scratch

matrix1 = np.array([[2, 4], [5,7]])
matrix2 = np.array([[6,8],[9,11]])

index_0_0 = (matrix1[0,0] * matrix2[0,0]) + (matrix1[0,1] * matrix2[1,0])
index_0_1 = (matrix1[0,0] * matrix2[0,1]) + (matrix1[0,1] * matrix2[1,1])
index_1_0 = (matrix1[1,0] * matrix2[0,0]) + (matrix1[1,1] * matrix2[1,0])
index_1_1 = (matrix1[1,0] * matrix2[0,1]) + (matrix1[1,1] * matrix2[1,1])

print(index_0_0)
print(index_0_1)
print(index_1_0)
print(index_1_1)

product1 = np.array([[index_0_0, index_0_1], [index_1_0, index_1_1]])
print(product1)

product2 = np.dot(matrix1, matrix2)
print(product2)
```
